```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('logo.png')

plt.figure()
# OpenCV image channel is BGR so we flip the channels to RGB
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
M = img.shape[1]
N = img.shape[0]
img = cv2.rectangle(img, (int(M/2-10),int(N/2-10)),(int(M/2+10),int(N/2+10)),(255,0,0),2)
font = cv2.FONT_HERSHEY_SIMPLEX

img = cv2.putText(img,"Prelab7",(80,35),font,1,(0,0,255),2)

plt.figure()
# OpenCV image channel is BGR so we flip the channels to RGB
plt.imshow(img[:,:,::-1])
plt.show()
```
