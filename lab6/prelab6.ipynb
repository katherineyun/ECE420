```python
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
from skimage import exposure
import copy
%matplotlib inline

# Implement this function
def imadd(pic,brightness):
    # Add brightness to each pixel
    brighten = pic + brightness
    return brighten

# Read the image
cameraman_origin = misc.imread('cameraman.tif')
# Create a copy of the origina image for us to manipulate
cameraman_bright = copy.deepcopy(cameraman_origin)
cameraman_bright1 = copy.deepcopy(cameraman_origin)

# Call imadd to perform enhancement
cameraman_bright = imadd(cameraman_bright,50)
cameraman_bright1 = imadd(cameraman_bright,300)
print(len(np.unique(cameraman_bright)))
# Show the results
fig_cam_origin = plt.figure(1)
fig_cam_origin.suptitle('Original Image')
plt.imshow(cameraman_origin,cmap='gray',vmin = 0, vmax = 255)
fig_cam_bright = plt.figure(2)
fig_cam_bright.suptitle('Brightened Image,50')
plt.imshow(cameraman_bright,cmap='gray',vmin = 0, vmax = 255)
fig_cam_bright = plt.figure(3)
fig_cam_bright.suptitle('Brightened Image,300')
plt.imshow(cameraman_bright1,cmap='gray',vmin = 0, vmax = 255)
plt.show()

eco_origin = misc.imread('eco.tif')
# Create a copy of the origina image for us to manipulate
eco_bright = copy.deepcopy(eco_origin)
eco_bright1 = copy.deepcopy(eco_origin)
# Call imadd to perform enhancement
eco_bright = imadd(eco_bright,50)
eco_bright1 = imadd(eco_bright,300)

# Show the results
fig_eco_origin = plt.figure(1)
fig_eco_origin.suptitle('Original Image,50')
plt.imshow(eco_origin,cmap='gray',vmin = 0, vmax = 255)
fig_eco_origin = plt.figure(2)
fig_eco_origin.suptitle('Original Image,50')
plt.imshow(eco_bright,cmap='gray',vmin = 0, vmax = 255)
fig_eco_bright = plt.figure(3)
fig_eco_bright.suptitle('Brightened Image,300')
plt.imshow(eco_bright1,cmap='gray',vmin = 0, vmax = 255)
plt.show()
```

```python
import numpy
from scipy import signal

def gengaussian(size=5,sigma=3.0):
    if size%2==0 or size<2:
        print('Size Not Valid')
        return None
    kernel = numpy.zeros((size,size))
    for x in range(size):
        for y in range(size):
            kernel[x][y] = numpy.exp(-((x-(size-1)/2)**2 \
                           +(y-(size-1)/2)**2)/(2*sigma**2))
    kernel = kernel / numpy.sum(kernel)
    return kernel

# Read Image and Display
kitten_origin = misc.imread('kitten.png')
# Create a copy of the origina image for us to manipulate
kitten_blur = copy.deepcopy(kitten_origin)
kitten_blur1 = copy.deepcopy(kitten_origin)

# Generate Kernel
kernel = gengaussian(5)
# Apply Convolution Here!
for i in range(3):
    kitten_blur[:,:,i] = signal.convolve2d(kitten_blur[:,:,i],kernel,mode = "same")
    
kernel = gengaussian(9)
for i in range(3):
    kitten_blur1[:,:,i] = signal.convolve2d(kitten_blur1[:,:,i],kernel,mode = "same")

# Display Results
fig_kitten_origin = plt.figure(1)
fig_kitten_origin.suptitle('Original Kitten.png', fontsize=14, fontweight='bold')
plt.imshow(kitten_origin,vmin = 0, vmax = 255)
fig_kitten_blur = plt.figure(2)
fig_kitten_blur.suptitle('Blurred Kitten.png, alpha = 5', fontsize=14, fontweight='bold')
plt.imshow(kitten_blur,vmin = 0, vmax = 255)
fig_kitten_blur = plt.figure(3)
fig_kitten_blur.suptitle('Blurred Kitten.png, alpha = 9', fontsize=14, fontweight='bold')
plt.imshow(kitten_blur1,vmin = 0, vmax = 255)
plt.show()
```

#### Answer:
The image becomes more blurry as the alpha value increases.
